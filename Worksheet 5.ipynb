```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
d_diam = sns.load_dataset('diamonds')
d_diam
```

```python
d_ice = sns.load_dataset('seaice')
d_ice
```

```python
d_diamFinal = sns.scatterplot(x = "carat", y = "price", data = d_diam, hue = 'cut')
d_diamFinal = d_diamFinal.set(title = "The price of diamonds", xlabel = "Weight (carats)", ylabel = "price ($)")
```

```python
plt.xticks = ([0, 3000, 6000, 9000, 12000], ["1980", "1992", "2000", "2008", "2016"])

plt.plot(d_ice["Extent"], color = "purple", linestyle = "--" )

plt.title("Change in Arctic sea ice thickness over time", fontsize=12, y=1.03)

plt.xlabel("Year", fontsize=14, labelpad=15)

plt.ylabel("Thickness (m)", fontsize=14, labelpad=15)




```
